## Calculator using Tikinter 

```python
from tkinter import *
import math as m
root =Tk()
root.title("Calculator")
e = Entry(root, width=25,font=("ariel",18), relief=RIDGE, fg="White", bg="Black",justify="right")
e.grid(row=0, column=0, columnspan=5, padx=10, pady=15,ipady=25 )
root.configure(bg="#ebfcfb")

def click(to_print):
    old=e.get()
    e.delete(0, END)
    e.insert(0, old+to_print)
    return

def sc(event):
    key=event.widget
    text=key['text']
    no=e.get()
    result=''

def clear(): 
    e.delete(0, END)  
    return

def backspace():
    current=e.get()
    length=len(current)-1
    e.delete(length, END)
    
def evaluate():
    ans=e.get()
    ans=eval(ans)
    e.delete(0, END)  
    e.insert(0, ans)
    
def Exit_Func():
    root.destroy()
def sc(event):
    key=event.widget
    text=key['text']
    no=e.get()
    result=''
    if text=='tan':
        result=str(m.tan(float(no)))
    if text=='sin':
        result=str(m.sin(float(no)))
    if text=='cos':
        result=str(m.cos(float(no)))
    if text=='X² ':
        result=str((float(no))*(float(no)))
    if text=='X³ ':
        result=str((float(no))*(float(no))*(float(no)))
    if text=='X!':
        result=str(m.factorial(float(no)))
    if text=='√X':
        result=str(m.sqrt(float(no)))
    if text=='%':
        result=str((float(no)/100))
    
    e.delete(0, END)   
    e.insert(0, result)

    
def on_Button1(e):
    btn_one['bg']="#a2b0af"
def on_Button2(e):
    btn_two['bg']="#a2b0af"
def on_Button3(e):
    btn_three['bg']="#a2b0af"
def on_Button4(e):
    btn_four['bg']="#a2b0af"
def on_Button5(e):
    btn_five['bg']="#a2b0af"
def on_Button6(e):
    btn_six['bg']="#a2b0af"
def on_Button7(e):
    btn_seven['bg']="#a2b0af"
def on_Button8(e):
    btn_eight['bg']="#a2b0af"
def on_Button9(e):
    btn_nine['bg']="#a2b0af"
def on_Buttonzero(e):
    btn_zero['bg']="#a2b0af"
def on_Buttondot(e):
    btn_dot['bg']="#a2b0af"
    
def off_Button1(e):
    btn_one['bg']="#f0f0f0"
def off_Button2(e):
    btn_two['bg']="#f0f0f0"
def off_Button3(e):
    btn_three['bg']="#f0f0f0"
def off_Button4(e):
    btn_four['bg']="#f0f0f0"
def off_Button5(e):
    btn_five['bg']="#f0f0f0"
def off_Button6(e):
    btn_six['bg']="#f0f0f0"
def off_Button7(e):
    btn_seven['bg']="#f0f0f0"
def off_Button8(e):
    btn_eight['bg']="#f0f0f0"
def off_Button9(e):
    btn_nine['bg']="#f0f0f0"
def off_Buttonzero(e):
    btn_zero['bg']="#f0f0f0"
def off_Buttondot(e):
    btn_dot['bg']="#f0f0f0"

def on_Buttonadd(e):
    btn_add['bg']="#c6d9c3"
def on_Buttonsub(e):
    btn_sub['bg']="#c6d9c3"
def on_Buttondiv(e):
    btn_div['bg']="#c6d9c3"
def on_Buttonmul(e):
    btn_mul['bg']="#c6d9c3"
def on_Buttonequal(e):
    btn_equal['bg']="#c6d9c3"

def off_Buttonadd(e):
    btn_add['bg']="#ebfffd"
def off_Buttonsub(e):
    btn_sub['bg']="#ebfffd"
def off_Buttondiv(e):
    btn_div['bg']="#ebfffd"
def off_Buttonmul(e):
    btn_mul['bg']="#ebfffd"
def off_Buttonequal(e):
    btn_equal['bg']="#CCEDFF"

def on_Buttonsin(e):
    btn_sin['bg']="#c6d9c3"
def on_Buttoncos(e):
    btn_cos['bg']="#c6d9c3"
def on_Buttontan(e):
    btn_tan['bg']="#c6d9c3"
def on_Buttonbackspace(e):
    btn_backspace['bg']="#fc5151"
def on_Buttonclear(e):
    btn_clear['bg']="#fc5151"
def on_Buttonsquare(e):
    btn_square['bg']="#c6d9c3"
def on_Buttoncube(e):
    btn_cube['bg']="#c6d9c3"
def on_Buttonfact(e):
    btn_fact['bg']="#c6d9c3"
def on_Buttonroot(e):
    btn_root['bg']="#c6d9c3"
def on_Buttonpercentage(e):
    btn_percentage['bg']="#c6d9c3"
    
def off_Buttonsin(e):
    btn_sin['bg']="#ebfffd"
def off_Buttoncos(e):
    btn_cos['bg']="#ebfffd"
def off_Buttontan(e):
    btn_tan['bg']="#ebfffd"
def off_Buttonbackspace(e):
    btn_backspace['bg']="#ebfffd"
def off_Buttonclear(e):
    btn_clear['bg']="#ebfffd"
def off_Buttonsquare(e):
    btn_square['bg']="#ebfffd"
def off_Buttoncube(e):
    btn_cube['bg']="#ebfffd"
def off_Buttonfact(e):
    btn_fact['bg']="#ebfffd"
def off_Buttonroot(e):
    btn_root['bg']="#ebfffd"
def off_Buttonpercentage(e):
    btn_percentage['bg']="#ebfffd"
    
    
btn_sin=Button(root,text="sin",padx=25.5,pady=20,borderwidth=0,bg="#ebfffd",fg="#000000")
btn_sin.bind("<Button-1>",sc)   
btn_sin.grid(row=1,column=0)
btn_sin.bind('<Enter>',on_Buttonsin)
btn_sin.bind('<Leave>',off_Buttonsin)

btn_cos=Button(root,text="cos",padx=24,pady=20,borderwidth=0,bg="#ebfffd",fg="#000000")
btn_cos.bind("<Button-1>",sc)
btn_cos.grid(row=1,column=1)
btn_cos.bind('<Enter>',on_Buttoncos)
btn_cos.bind('<Leave>',off_Buttoncos)

btn_tan=Button(root,text="tan",padx=23.8,pady=20,borderwidth=0,bg="#ebfffd",fg="#000000")
btn_tan.bind("<Button-1>",sc)   
btn_tan.grid(row=1,column=2)
btn_tan.bind('<Enter>',on_Buttontan)
btn_tan.bind('<Leave>',off_Buttontan)

exit_button=Button(root,text="Exit",command=Exit_Func,bg="red", fg="White",font=('italic',10),borderwidth=0,cursor="hand2")
exit_button.grid(row=1,column=3)

btn_square=Button(root,text="X² ",padx=25.5,pady=20,borderwidth=0,bg="#ebfffd",fg="#000000")
btn_square.bind("<Button-1>",sc)   
btn_square.grid(row=2,column=0)
btn_square.bind('<Enter>',on_Buttonsquare)
btn_square.bind('<Leave>',off_Buttonsquare)

btn_cube=Button(root,text="X³ ",padx=25.5,pady=20,borderwidth=0,bg="#ebfffd",fg="#000000")
btn_cube.bind("<Button-1>",sc)   
btn_cube.grid(row=2,column=1)
btn_cube.bind('<Enter>',on_Buttoncube)
btn_cube.bind('<Leave>',off_Buttoncube)

btn_fact=Button(root,text="X!",padx=28,pady=20,borderwidth=0,bg="#ebfffd",fg="#000000")
btn_fact.bind("<Button-1>",sc)   
btn_fact.grid(row=2,column=2)
btn_fact.bind('<Enter>',on_Buttonfact)
btn_fact.bind('<Leave>',off_Buttonfact)

btn_root=Button(root,text="√X",padx=37,pady=20,borderwidth=0,bg="#ebfffd",fg="#000000")
btn_root.bind("<Button-1>",sc)   
btn_root.grid(row=2,column=3)
btn_root.bind('<Enter>',on_Buttonroot)
btn_root.bind('<Leave>',off_Buttonroot)

btn_clear=Button(root,text="C",padx=27,pady=20,command=clear,relief=FLAT,bg="#ebfffd",activebackground="#ff0000",fg="#000000")
btn_clear.grid(row=3,column=0)
btn_clear.bind('<Enter>',on_Buttonclear)
btn_clear.bind('<Leave>',off_Buttonclear)

btn_backspace=Button(root,text="⌫ ",padx=22.3,pady=20,command=backspace,relief=FLAT,activebackground="#ff1c1c",bg="#ebfffd",fg="#000000")
btn_backspace.grid(row=3,column=1)
btn_backspace.bind('<Enter>',on_Buttonbackspace)
btn_backspace.bind('<Leave>',off_Buttonbackspace)

btn_percentage=Button(root,text="%",padx=28,pady=21.5,borderwidth=0,bg="#ebfffd",fg="#000000")
btn_percentage.bind("<Button-1>",sc)   
btn_percentage.grid(row=3,column=2)
btn_percentage.bind('<Enter>',on_Buttonpercentage)
btn_percentage.bind('<Leave>',off_Buttonpercentage)

btn_div=Button(root,text="/",padx=39.5,pady=20,command=lambda:click('/'),relief=FLAT,bg="#ebfffd",fg="#000000")
btn_div.grid(row=3,column=3)
btn_div.bind('<Enter>',on_Buttondiv)
btn_div.bind('<Leave>',off_Buttondiv)

btn_seven = Button(root, text="7", padx=30,pady=20,borderwidth=0,  fg="black",command=lambda: click("7"))
btn_seven.grid(row=4, column=0)
btn_seven.bind('<Enter>',on_Button7)
btn_seven.bind('<Leave>',off_Button7)

btn_eight = Button(root, text="8",padx=30,pady=20,borderwidth=0,  fg="black",command=lambda: click("8"))
btn_eight.grid(row=4, column=1)
btn_eight.bind('<Enter>',on_Button8)
btn_eight.bind('<Leave>',off_Button8)

btn_nine = Button(root, text="9", padx=30,pady=20,borderwidth=0,   fg="black",command=lambda: click("9"))
btn_nine.grid(row=4, column=2)
btn_nine.bind('<Enter>',on_Button9)
btn_nine.bind('<Leave>',off_Button9)

btn_mul= Button(root, text="* ", padx=40, pady=20,borderwidth=0,bg="#ebfffd",fg="black",command=lambda: click("*"))
btn_mul.grid(row=4, column=3)
btn_mul.bind('<Enter>',on_Buttonmul)
btn_mul.bind('<Leave>',off_Buttonmul)

btn_four = Button(root, text="4", padx=30,pady=20,borderwidth=0,  fg="black",command=lambda: click("4"))
btn_four.grid(row=5, column=0)
btn_four.bind('<Enter>',on_Button4)
btn_four.bind('<Leave>',off_Button4)

btn_five = Button(root, text="5",padx=30,pady=20,borderwidth=0,  fg="black",command=lambda: click("5"))
btn_five.grid(row=5, column=1)
btn_five.bind('<Enter>',on_Button5)
btn_five.bind('<Leave>',off_Button5)

btn_six = Button(root, text="6",padx=30,pady=20, borderwidth=0, fg="black",command=lambda: click("6"))
btn_six.grid(row=5, column=2)
btn_six.bind('<Enter>',on_Button6)
btn_six.bind('<Leave>',off_Button6)

btn_sub = Button(root, text="- ", padx=40, pady=20,borderwidth=0,bg="#ebfffd",  fg="black",command=lambda: click("-"))
btn_sub.grid(row=5, column=3)
btn_sub.bind('<Enter>',on_Buttonsub)
btn_sub.bind('<Leave>',off_Buttonsub)

btn_one = Button(root, text="1",padx=30,pady=20,borderwidth=0,   fg="black",command=lambda: click("1"))
btn_one.grid(row=6, column=0)
btn_one.bind('<Enter>',on_Button1)
btn_one.bind('<Leave>',off_Button1)

btn_two = Button(root, text="2",padx=30,pady=20,borderwidth=0,   fg="black",command=lambda: click("2"))
btn_two.grid(row=6, column=1)
btn_two.bind('<Enter>',on_Button2)
btn_two.bind('<Leave>',off_Button2)

btn_three = Button(root, text="3", padx=30,pady=20,borderwidth=0, fg="black",command=lambda: click("3"))
btn_three.grid(row=6, column=2)
btn_three.bind('<Enter>',on_Button3)
btn_three.bind('<Leave>',off_Button3)

btn_add = Button(root, text="+", padx=40, pady=20,borderwidth=0,bg="#ebfffd",  fg="black",command=lambda: click("+"))
btn_add.grid(row=6, column=3)
btn_add.bind('<Enter>',on_Buttonadd)
btn_add.bind('<Leave>',off_Buttonadd)

btn_dot = Button(root, text=" .", padx=30,pady=20,borderwidth=0,  fg="black",command=lambda: click("."))
btn_dot.grid(row=7, column=0)
btn_dot.bind('<Enter>',on_Buttondot)
btn_dot.bind('<Leave>',off_Buttondot)

btn_zero = Button(root, text="0", padx=30,pady=20,borderwidth=0,  fg="black",command=lambda: click("0"))
btn_zero.grid(row=7, column=1)
btn_zero.bind('<Enter>',on_Buttonzero)
btn_zero.bind('<Leave>',off_Buttonzero)

btn_equal = Button(root, text=" = ", padx=81, pady=20,borderwidth=0, bg="#CCEDFF",command=lambda: evaluate())
btn_equal.grid(row=7, column=2, columnspan=3)
btn_equal.bind('<Enter>',on_Buttonequal)
btn_equal.bind('<Leave>',off_Buttonequal)


root.mainloop()
```
